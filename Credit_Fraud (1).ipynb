# Applying PCA

```python
#Applying PCA and plotting scree plots to make inferences

import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Step 1: Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Step 2: Apply PCA
pca = PCA()
pca.fit(scaled_data)
explained_variance = pca.explained_variance_ratio_

# Step 3: Plot the scree plot
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o', linestyle='-')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.grid(True)
plt.show()

# Step 4: Cumulative variance plot (Optional but helpful for interpretation)
cumulative_variance = np.cumsum(explained_variance)
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='-')
plt.title('Cumulative Variance Explained')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Variance Explained')
plt.grid(True)
plt.show()
```

```python
# Step 3: Plot the scree plot as a bar graph
plt.figure(figsize=(10, 6))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.7, color='blue')
plt.title('Scree Plot')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.xticks(range(1, len(explained_variance) + 1))  # Set x-ticks to match principal components
plt.grid(axis='y')
plt.show()
```

```python
#We see that each Principal Component contributes to explaining some amount of the variation of the data, 
#around 22 PCs explain 80% of the variation in the dataset
#around 26 PCs explain 90% of the variation in the dataset
```

```python
#Applying random forest classifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


# Step 1: Separate features and target variable
X = df.drop('Class', axis=1)
y = df['Class']

# Step 4: Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 5: Train a Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Step 6: Make predictions
y_pred = rf_model.predict(X_test)

# Step 7: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
#Accuracy is high as expected in case of imbalanced data
#but from the confusion matrix we can identify that only 109/136 fraud transactions were classified correctly
```

# Applying SMOTE for imbalanced dataset

```python
from imblearn.over_sampling import SMOTE

# Step 5: Apply SMOTE to upsample the minority class in the training set
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Step 6: Train a Random Forest model on the balanced data
rf_model_smote = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model_smote.fit(X_train_smote, y_train_smote)

# Step 7: Make predictions
y_pred = rf_model_smote.predict(X_test)

# Step 8: Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
#after applying smote, we see improvement in flagging fraud transactions 119/136
#although we also see the precision decrease, but a very meaningful and significant increase on recall
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
import numpy as np
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# Step 1: Separate features and target variable
X = df.drop('Class', axis=1)
y = df['Class']
X_scaled = scaler.fit_transform(X)

# Step 3: Split the data into training and testing sets using stratification
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42, stratify=y)

# Step 5: Apply SMOTE to upsample the minority class in the training set
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Step 4: Use KNN to determine neighbor class proportions
knn = KNeighborsClassifier(n_neighbors=3)  # You can adjust the number of neighbors as needed
knn.fit(X_train_smote, y_train_smote)

# Calculate class probabilities for the training and test sets based on KNN
train_probs_knn = knn.predict_proba(X_train_smote)
test_probs_knn = knn.predict_proba(X_test)

# Add KNN probabilities as new features to the training and testing datasets
X_train_with_knn_probs = np.hstack((X_train_smote, train_probs_knn))
X_test_with_knn_probs = np.hstack((X_test, test_probs_knn))

# Step 5: Train a Random Forest model on the updated dataset with engineered features
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_with_knn_probs, y_train_smote)

# Step 6: Make predictions with the updated Random Forest model
y_pred = rf_model.predict(X_test_with_knn_probs)

# Step 7: Evaluate the Random Forest model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
